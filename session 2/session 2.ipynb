```python
#reusing the class I wrote in the previous session

import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack

%matplotlib inline

plt.style.use('ggplot')

class CHO(): #CHO - coupled harmonic oscillator
    
    def __init__(self ,N , m , k , t_max):
        
        self.m = m #mass in kg
        self.k = k #spring constant k 
        self.N = N #number of particles
        self.dt = 0.01 # dt in s
        self.xn = [[] for x in range(N)] #displacement arrays of N particles
        self.vn = [[] for x in range(N)] #velocity arrays of N particles
        self.KE = [] #kinetic energy in joules
        self.PE = [] #potential energy in joules
        self.TE = [] #total energy in joules 
        self.t_max = t_max #maximum time 
        self.t_s = np.arange(0 , self.t_max , self.dt)
        self.t = 0
        self.current_q = [0 , 0.1] #same initial parameters as for 3 coupled oscillator , spring displacements
        
        for _ in range(N-1):
            
            self.current_q.append(0)
        
        self.current_q.append(0)
        
        self.current_v = [0 for x in range(N)] 
    
    def one_particle(self , s_no): #calculates x , v , a for one particle
        
        a_particle = (self.k*(self.current_q[s_no+1] - 2*self.current_q[s_no] + self.current_q[s_no-1]))/self.m #calculating acceleration
        
        v_new = self.current_v[s_no-1] + 0.5*a_particle*self.dt #calculating new velocity
        self.current_v[s_no-1] = v_new
        self.vn[s_no-1].append(v_new)
        
        x_new = self.current_q[s_no] + v_new*self.dt #calculating new displacement
        self.current_q[s_no] = x_new
        self.xn[s_no-1].append(x_new)
    
    def energy(self , t):
        
        pe_new = 0
        
        for i in range(0 , self.N):
            
            x_spring = self.current_q[i+1] - self.current_q[i]
            
            pe_new += 0.5*self.k*x_spring*x_spring
            
        if t==0:
            
            self.TE.append(pe_new)
            self.PE.append(pe_new)
            self.KE.append(0)
        
        else:
            
            ke_new = self.TE[0] - pe_new
            self.TE.append(ke_new+pe_new)
            self.KE.append(ke_new)
            self.PE.append(pe_new)
    
    def VerletAlgorithm(self):
        
        while self.t <= self.t_max:
            
            for i in range(1,self.N+1):
                
                self.one_particle(i)
            
            self.energy(self.t)
            
            self.t+=self.dt
    
    def __GenerateLegend(self): #private class method to generate the legend for each plot
        
        self.Legend = [str(n) for n in range(self.N+1)]
    
    def PlotDisplacement(self): #function to plot displacements of each particle
        
        plt.figure(figsize = (30 , 10))
        plt.title("displacement of each particle over time")
        
        self.__GenerateLegend()
        
        for x in self.xn:
            
            plt.plot(self.t_s , x)
        
        plt.xlabel("time in seconds")
        plt.ylabel("displacement in meters")
        plt.legend(self.Legend)
        plt.show()
    
    def PlotVelocity(self): #function to plot velocities of each particle
        
        plt.figure(figsize = (30 , 10))
        plt.title("velocity of each particle over time")
        
        self.__GenerateLegend()
        
        for v in self.vn:
            
            plt.plot(self.t_s , v)
        
        plt.xlabel("time in seconds")
        plt.ylabel("displacement in meters")
        plt.legend(self.Legend)
        plt.show()
        
    def PlotEnergy(self):
        
        plt.figure(figsize = (30,10))
        plt.title("energies of the system over time")
        
        plt.plot(self.t_s , self.KE)
        plt.plot(self.t_s , self.PE)
        plt.plot(self.t_s , self.TE)
        
        plt.xlabel("time in seconds")
        plt.ylabel("energy in joules")
        
        plt.legend(["kinetic energy" , "potential energy" , "total energy"])
    
        plt.show()
```

```python
#oscillator with 2 masses 

o = CHO(6 , 0.1 , 1 , 100)
o.VerletAlgorithm()
o.PlotDisplacement()
```

```python
X = fftpack.fft(o.xn[0])
freqs = fftpack.fftfreq(len(o.xn[0]) , d = o.dt)
print(X[:])
print(max(X))

fig , ax = plt.subplots()

plt.stem(freqs , np.abs(X))
plt.figure(figsize=(30,10))
plt.show()
```
